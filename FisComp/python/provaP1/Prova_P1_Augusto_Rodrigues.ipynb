### Listas:
Listas são estruturas de dados compostas por indice (começando pelo zero) e valor, pode receber valores repetidos e de multiplos tipos. É mutavel e delimitada com colchetes.

Principalmente usada para coleções de itens que podem ser modificados, como uma lista numeros aleatorios para simular uma jogada de dados, como no exemplo abaixo.

```python
from numpy import random


# Exemplos com listas
minha_lista = [1,2,3,4,5,6,7,7]

#Imprimir primeiro elemento na tela
print(minha_lista[0])

#Imprimir ultimo elemento na tela
print(minha_lista[-1])

#Numeros aleatorios
#simula 10 jogadas de dados de 6 faces e adiciona na lista
rand_nums = []
for i in range(11):
    rand_nums.append(random.randint(1,7))

print(rand_nums)
```

### Tuplas
Tuplas são estruturas de dados compostas por indice (começando pelo zero) e valor, podem receber valores repetidos e de multiplos tipos, é imutavel e pode ou nao ser delimitada por parenteses.

Principal uso é para manter elementos pré determinados salvos e, por exemplo, fazer comparações, como no exemplo abaixo que verifica se o numero da tupla é par ou não.

```python
# Exemplos com tuplas
minha_tupla = (1,2,3,4,5,6,7,7)

#Imprimir primeiro elemento na tela
print(minha_tupla[0])

#Imprimir ultimo elemento na tela
print(minha_tupla[-1])


for i in range(len(minha_tupla)):
    if minha_tupla[i] % 2 == 0:
        print('par')
    else:
        print('impar')
```

### Conjuntos

Conjuntos são estruturas de dados compostas por indice (começando pelo zero) e valor, não podem receber valores repetidos porém pode receber multiplos tipos, é mutavel e delimitado por chaves.

Pode ser usado como um filtro, como no exemplo abaixo, ou também com operações de conjunto.

```python
#Exemplos com conjuntos
conj_1 = {1,2,3,4,5,6,7}
conj_2 = {5,6,7,8,9,10,11}

#Imprimir conjunto tela
print(conj_1)

#imprimir a interceção dos conjuntos na tela
print(conj_1 & conj_2)

#imprimir a união dos conjuntos na tela
print(conj_1 | conj_2)

#filtrar elementos de uma lista
lista_conj = [1,2,11,13,4,5,30]
lista_filtrada = []

for j in range(len(lista_conj)):
    if lista_conj[j] not in conj_1:
        lista_filtrada.append(lista_conj[j])   
print(lista_filtrada)
```

```python
'''Erro 2: o operador para somar uma variavel com ela mesma é variavel += numero, no código está variavel =+ numero'''

def calcular_media_final(lista_numeros):
    soma = 0
    for i in range(len(lista_numeros)):
        soma += lista_numeros[i]
    media = soma / len(lista_numeros)
    return media
numeros = [10,20,30,40]
print("A média é:", calcular_media_final(numeros))
```

```python
def sinal(tempo):
    cores = []
    for i in range(1,tempo+1):
        if i % 2 == 0 and i % 5 == 0:
           cores.append({i:'vermelho'})
           break
        if i %2 == 0:
           cores.append({i:'Verde'})
        if i %5 == 0:
           cores.append({i:'Amarelo'})
    return cores
            


N = int(input('Digite um numero'))

cores = sinal(N)

for i in range(len(cores)):
        for n, cor_sinal in cores[i].items():
            print(f"{n} : {cor_sinal}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pos_x(t, theta, v0):
    theta = theta*np.pi/180
    return v0 * np.cos(theta) * t

def pos_y(t, theta, v0):
    g = 9.8
    theta = theta*np.pi/180
    return v0*np.sin(theta)*t - ((1/2)*(g*(t**2)))

tempos = np.arange(0, 9, 0.1)

posicao_x = pos_x(tempos, 60, 50)
posicao_y = pos_y(tempos, 60, 50)

plt.plot(posicao_x,posicao_y, linewidth=1.0, color = 'red', label='Posição')
plt.xlabel('Posição em x')
plt.ylabel('Posição em y')
plt.title('Lançamento de projétil',fontsize=18)
plt.legend()
plt.grid() 
plt.ylim(0) 
plt.show()
```
