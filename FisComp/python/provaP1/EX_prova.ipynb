Questão 1 - (1.5 pontos)

Explique a diferença entre listas, tuplas e dicionários em Python. Apresente exemplos de código que demonstram a manipulação de cada uma uma destas estruturas de dados.

```python
'''
Listas são estruturas de dados compostas por um indice e um valor, mutaveis e delimitadas por colchetes
'''
lista = [1,3,4,5,6]
print(lista[1])
'''
Tuplas são estruturas de dados também compostas por indice e valor, porem sao imutaveis e delimitadas por parenteses
'''

tupla = (1,2,3,4)
print(tupla[3])
'''
Dicionarios são estruturas de dados compostos por chave e valor, são delimitados por chaves, e são mutaveis
'''

dicionario = {
    'chave': 'valor1',
    'chave2': 'valor2'
}
print(dicionario['chave'])
```

#### Questão 2 - (3.5 pontos)
a) Escreva uma função Python que calcule a soma dos termos da seguinte série
    
$$\sum_{n=0}^N x^{2n}$$

A função deve receber como entrada o valor de $x$ e do número total de termos $N$.

b) Usando $x=0.8$, calcule essa soma para N = 50, 100, 150 e determine o erro com relação ao resultado analítico da série (para $N=\infty$), dado por
$$\frac{1}{1-x^2}$$

c) Qual sua conclusão sobre a evolução do erro com $N$?

```python
def somatorio(x,N):
    total = 0
    for i in range(N+1):
        total = total+(x**(2*i))
    return total

calcula_resultado_analitico = lambda x: 1/(1-(x**2))

resultado_analitico = calcula_resultado_analitico(0.8)
print(resultado_analitico)

resultado50 = somatorio(0.8,50)
resultado100 = somatorio(0.8,100)
resultado150 = somatorio(0.8,150)

print(resultado50)
print(resultado100)
print(resultado150)
```

#### Questão 3 - (2.5 pontos)



O fluxograma abaixo representa um algoritmo para o cálculo do fatorial de um número natural n. Codifique o algoritmo abaixo em uma função Python que recebe como entrada o valor de N e retorna o seu fatorial. Teste seu algoritmo para N=5.

algoritmo_fatorial.png




```python
def fatorial(N):
    fat = 1
    i=1
    while i <= N:
        fat = fat * i
        i += 1
    else:
        return fat

print(fatorial(5))
```

#### Questão 4 - (2.5 pontos)

Considere a função

$g(x, n) = \frac{sen(x)}{x^n}$

(a) Crie uma função chamada g_func(x,n), que retorna o valor de $g(x, n)$

(b) Use esta função para plotar $sen(x)/x$, $sen(x)/x^2$ e $sen(x)/x^3$ em um
mesmo gráfico para −10 < x < 10.

```python
import numpy as np
import matplotlib.pyplot as plt

def g_func(x,n):
    
    return np.sin(x)/x**n

intervalo = np.delete(np.arange(-10, 10.1, 0.1), 100)


resultado_linear = g_func(intervalo, 1)
resultado_quadrado = g_func(intervalo, 2)
resultado_cubo = g_func(intervalo, 3)


print(resultado_linear)
plt.plot(intervalo, resultado_linear)
plt.plot(intervalo, resultado_quadrado)
plt.plot(intervalo, resultado_cubo)
plt.show()
```
