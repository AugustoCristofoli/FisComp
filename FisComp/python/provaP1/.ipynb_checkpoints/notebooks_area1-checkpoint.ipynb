##Variáveis

**O que são Variáveis?**

Variáveis são usadas para armazenar dados. Em Python, não precisamos declarar o tipo da variável explicitamente — o interpretador infere o tipo automaticamente com base no valor atribuído. Isso é o que chamamos de tipagem dinâmica.

Em Python, usamos o símbolo = para atribuir um valor a uma variável.

````python
# Atribuindo valores a variáveis
x = 10  # Atribuindo um número inteiro à variável 'x'
mensagem = "Olá, Mundo!"  # Atribuindo uma string à variável 'mensagem'

# Exibindo os valores das variáveis
print(x)
print(mensagem)
````

**Regras para Nomes de Variáveis**

Os nomes de variáveis podem ser escolhidos livremente, mas precisam seguir certas regras:

* Podem conter letras (a-z, A-Z), dígitos (0-9) e o caractere sublinhado (_).
* Não podem começar com dígitos.
* Não podem usar palavras reservadas do Python (como if, else, for, etc.).
* Devem ser descritivos para facilitar a leitura do código.

Exemplos de nomes válidos:





````python
nome = "João"
idade = 25
altura_cm = 170
velocidade_maxima = 100
````

Exemplos de nomes inválidos:



````python
1nome = "João"  # Nomes de variáveis não podem começar com números
nome! = "Maria"  # Não pode conter caracteres especiais (!)
````

Se tentarmos acessar uma variável antes de atribuir um valor a ela, o Python retornará um erro do tipo NameError.

````python
# Tentando acessar uma variável que não foi definida
print(nome_inexistente)  # Isso resultará em um NameError
````

##Números

O Python suporta diferentes tipos numéricos, que são usados para realizar operações matemáticas:

  * Inteiros (int): Números inteiros
  
  Ex: -5, 0, 42
  * Números de ponto flutuante (float): Números que contêm casas decimais
  
  Ex: 3.14, -0.001, 2.5

  * Números complexos (complex): Números que incluem uma parte real e uma parte imaginária
  
  Ex: 3 + 2j



````python
# Inteiro
a = 10
print(type(a))  # <class 'int'>

# Float (número decimal)
b = 3.14
print(type(b))  # <class 'float'>

# Número complexo
c = 2 + 3j
print(type(c))  # <class 'complex'>
````

A função type() em Python é usado para verificar o tipo de dado de uma variável ou valor. Ela retorna o tipo do objeto passado como argumento.

**Operadores Algébricos**


 *   Adição (+): Soma dois valores
 *   Subtração (-): Subtrai o segundo valor do primeiro
 *   Multiplicação (*): Multiplica dois valores.
 *   Divisão (/): Divide o primeiro valor pelo segundo, sempre retornando um float.
 *   Divisão inteira (//): Divide o primeiro valor pelo segundo e retorna o resultado inteiro.
 *   Módulo (%): Retorna o resto da divisão.
 *   Exponenciação (**): Eleva o primeiro valor à potência do segundo.



````python
# Adição
soma = 5 + 3  # Resultado: 8
print(soma)

# Subtração
subtracao = 10 - 4  # Resultado: 6
print(subtracao)

# Multiplicação
multiplicacao = 7 * 3  # Resultado: 21
print(multiplicacao)

# Divisão (float)
divisao = 9 / 3  # Resultado: 3.0
print(divisao)

# Divisão inteira
divisao_inteira = 10 // 3  # Resultado: 3
print(divisao_inteira)

# Módulo
resto = 10 % 3  # Resultado: 1
print(resto)

# Exponenciação
potencia = 2 ** 3  # Resultado: 8
print(potencia)
````

Além do operador de atribuição (=), em Python temos operadores de atribuição compostos(+=, -=, *=, /=). Eles são usados para atualizar o valor de uma variável, somando (subtraindo/multiplicando/dividindo) um valor adicional à variável existente, e depois armazenando o resultado de volta na própria variável.

````python
a = 1
a += 3   # equivalente a = a + 3
print(a)
````

````python
b = 100
b /= 10 #equivalente b = b/10
print(b)
````

Assim como na matemática, Python segue uma ordem de prioridade para realizar operações algébricas. A ordem de operações pode é Exponenciação, Multiplicação e Divisão, Adição e Subtração.

Podemos usar parênteses para alterar a ordem padrão das operações e garantir que certos cálculos sejam realizados primeiro.

   

````python
resultado = 5 + 3 * 2  # Multiplicação ocorre primeiro: 5 + (3 * 2) = 11
print(resultado)  # Resultado: 11

resultado = (5 + 3) * 2  # Parênteses têm prioridade: (5 + 3) * 2 = 16
print(resultado)  # Resultado: 16

resultado = 2 ** 3 * 2  # Exponenciação ocorre primeiro: (2 ** 3) * 2 = 8 * 2 = 16
print(resultado)  # Resultado: 16

resultado = -3 ** 2  # Exponenciação ocorre primeiro: (3 ** 2) do que multiplicação (por -1)
print(resultado)  # Resultado: -9

resultado = (-3) ** 2
print(resultado)  # Resultado: 9
````

##Strings

Strings são sequências de caracteres, que podem ser descritas de diversas formas. Elas são delimitadas por aspas simples (' ') ou duplas (" ").



````python
# Usando aspas simples
saudacao = 'Olá, mundo!'

# Usando aspas duplas
mensagem = "Bem-vindo ao Python!"
````

**Operações básicas com Strings**

*Concatenar Strings*: é possível juntar (concatenar) strings usando o operador +.

````python
nome = "Alice"
saudacao = "Olá, " + nome + "!"
print(saudacao)  # Resultado: "Olá, Alice!"
````

*Repetir Strings*: o operador * permite repetir uma string.

````python
repetir = "Python! " * 3
print(repetir)  # Resultado: "Python! Python! Python! "
````

**Indexação de Strings**

As strings em Python são indexadas, ou seja, cada caractere tem uma posição associada, começando do índice 0. Podemos acessar os elementos de uma string pelo seu índice.

````python
palavra = 'python'

print(palavra[0]) #imprime o primeiro caractere da string.
print(palavra[1]) #imprime o segundo caractere da string.
print(palavra[-1]) #imprime o último caractere. Índices negativos contam de trás para a frente.
````

Strings podem ser 'fatiadas' (slicing). Devemos especificar o índice inicial (incluído) e o final (excluído), separados por ':'

````python
print(palavra[2:5])  #imprime do terceiro caractere (índice 2) ao quinto caractere (índice 4).
print(palavra[:3])   #quando índice inicial não é especificado, assume que é o primeiro (índice 0)
print(palavra[2:])   #quando índice final não é especificado, assume que vai até o último
````

**Strings são Imutáveis**

Em Python, strings são imutáveis, o que significa que uma vez criadas, seus valores não podem ser alterados diretamente. Para modificar uma string é preciso criar uma nova string.



````python
texto = "Olá"
# texto[0] = "B"  # Isso causaria um erro
novo_texto = "B" + texto[1:]  # Correto: criando uma nova string
print(novo_texto)  # Resultado: "Blá"
````

*Comprimento da String*: A função len() retorna o número de elementos em uma string.


````python
print(len(texto))  # Resultado: 3
````

Também podemos criar uma string vazia:

````python
txt = ''
````

##Listas


Uma lista é uma coleção ordenada e mutável de itens. As listas são muito flexíveis e podem armazenar diferentes tipos de dados, como números, strings e até outras listas.

Os valores em uma lista são delimitados por [ ] e separados por vírgulas.


````python
# Exemplo de uma lista de números
numeros = [1, 2, 3, 4, 5]

# Exemplo de uma lista com diferentes tipos de dados
misturada = [10, "Python", 3.14, True]
````

**Acessando Elementos da Lista (Indexação)**

As listas são indexadas, o que significa que cada elemento tem uma posição (índice).

A indexação dos elementos das listas segue o mesmo esquema apresentado acima para strings, ou seja, com os índices começam em 0.

````python
frutas = ["maçã", "banana", "laranja"]
print(frutas[0])  # Acessa o primeiro elemento: 'maçã'
print(frutas[-1])  # Acessa o último elemento: 'laranja'
````

**Modificando Elementos de Listas**

Ao contrário das strings, as listas são mutáveis, o que significa que é possível modificar seus elementos diretamente.

````python
frutas[1] = "uva"  # Substitui o elemento de índice 1 ('banana') por 'uva'
print(frutas)  # Resultado: ['maçã', 'uva', 'laranja']
````

**Operações Básicas com Listas**

*Concatenar Listas*:  é possível concatenar (juntar) duas ou mais listas usando o operador +

````python
outras_frutas = ["abacaxi", "manga"]
todas_frutas = frutas + outras_frutas
print(todas_frutas)  # Resultado: ['maçã', 'uva', 'laranja', 'abacaxi', 'manga']
````

*Adicionar Elementos*: podemos adicionar itens a uma lista usando o método append().

````python
todas_frutas.append("morango")  # Adiciona 'morango' ao final da lista
print(todas_frutas)  # Resultado: ['maçã', 'uva', 'laranja', 'abacaxi', 'manga', 'morango']
````

*Remover Elementos*: podemos remover itens usando o método remove().

````python
todas_frutas.remove("uva")  # Remove 'uva' da lista
print(todas_frutas)  # Resultado: ['maçã', 'laranja', 'abacaxi', 'manga', 'morango']
````

*Fatiamento (Slicing) de Listas*: podemos obter sublistas utilizando a notação de fatiamento (slicing). A sintaxe é similar ao fatiamento de strings.

````python
numeros = [1, 2, 3, 4, 5, 6]
print(numeros[1:4])  # Resultado: [2, 3, 4] (do índice 1 ao 3)
print(numeros[:3])   # Resultado: [1, 2, 3] (do início ao índice 2)
````

*Comprimento da Lista*: A função len() retorna o número de elementos em uma lista.

````python
print(len(numeros))  # Resultado: 6
````

## Condicionais e comando if

**Comparações**

Operadores de comparação em Python são usados para comparar dois valores ou variáveis e retornar um valor booleano (True ou False). Eles são essenciais para implementar lógica condicional em programas.

*Operadores de comparação:*

* == : igual a
* != : diferente de
* \> : maior que
* \>= : maior ou igual a
* < : menor que
* <= : menor ou igual a






````python
6 * 2 == 9 + 3  #True
````

````python
8*6.8 == 7.8*1.9  #False
````

````python
5 != 3  # True
````

````python
a = 5.5
b = 7.3
c = 9.8
a*2 > b  #True
````

*Operadores Booleanos:*

Os operadores booleanos são usados para combinar múltiplas expressões condicionais. Eles retornam um valor booleano baseado nos resultados combinados das expressões.

* and : Retorna True se todas as condições forem verdadeiras. Caso alguma seja falsa, retorna False.
* or : Retorna True se pelo menos uma das condições for verdadeira. Só retorna False se todas forem falsas.
* not : Inverte o valor booleano da condição. Se a condição for True, retorna False, e vice-versa.

````python
a = 5.5
b = 7.3
c = 9.8
(b > a) and (b < c)  #True
````

````python
(5 < 3) or (10 < 7)  # False
````

````python
not (5 < 3)  # True
````

*Operador is*

O operador is é usado para verificar se duas variáveis referenciam o mesmo objeto na memória. Ele não compara valores, mas sim a identidade do objeto. Se dois objetos ocupam o mesmo espaço na memória, is retorna True.

````python
a = [1, 2, 3]
b = a  # 'b' aponta para o mesmo objeto que 'a'
c = [1, 2, 3]  # 'c' é um objeto diferente com o mesmo conteúdo

a[0] = 10  # altera o primeiro elemento da lista a
print(a)  # [10, 2, 3] -> 'a' foi alterado
print(b)  # [10, 2, 3] -> 'b' também foi alterado, pois é igual a 'a'

print(a is b)  # True, pois 'b' referencia o mesmo objeto que 'a'
print(a is c)  # False, pois 'c' é um objeto diferente, apesar de ter o mesmo valor
````

Diferença entre is e ==

  * is verifica se dois objetos ocupam o mesmo local na memória (verifica a identidade).
  * == compara os valores dos objetos (verifica a igualdade).

*Operador in*

O operador in é usado para verificar se um elemento está presente em uma sequência, como listas e strings (e tuplas ou dicionários, que veremos nas próximas aulas). Ele retorna True se o elemento for encontrado, e False caso contrário.

````python
frutas = ['maçã', 'banana', 'laranja']

print('banana' in frutas)  # True, pois 'banana' está na lista
print('abacaxi' in frutas)  # False, pois 'abacaxi' não está na lista
````

## comando IF

O comando if funciona de forma a verificar se uma condição é verdadeira. Caso a mesma seja, o algoritmo executará um conjunto de ações. Caso contrário, o algoritmo poderá seguir para a próxima ação.

Em Python, o comando if é executado com a seguinte sintaxe:

      if condição:
        executa ação

Note a utilização de : ao final da condição que serve como um encerramento da verificação de todas as condições.

Note também que existe uma quebra de linha seguido de um espaçamento, que iremos chamar de indentação. A indentação é necessária devido a particularidades da linguagem Python, que utiliza a indentação para identificar um bloco de código.


````python
# Você pode votar nas eleições ?
idade = 10
print("Idade =",idade,"anos")
if idade >= 18:
  print("Você tem obrigação de votar.")
````

Mas e se for necessário identificar que qualquer resultado contrário à condição do if ?

Para isso utilizamos o comando else.

````python
# Você pode votar nas eleições ?
idade = 12
print("Idade =",idade,"anos")
if idade >= 18 :
  print("Você tem obrigação de votar.")
else:
  print("Você não tem idade para votar.")
````

E no caso de termos um meio termo, onde o valor da condição não é atingido, mas também não é um valor de saída? Em Python utilizamos o comando elif, que funciona de forma semelhante ao if.



````python
# Você pode votar nas eleições ?
idade = 17
print("Idade =",idade,"anos")
if idade >= 18:
  print("Você tem obrigação de votar.")
elif idade >= 16:
  print("O voto é opcional.")
else:
  print("Você não tem idade para votar.")
````

Seja consistente na indentação. Evite usar tecla tab. Use 4 espaços. A maioria dos editores farão a indentação automaticamente.

Os blocos são fechado através da 'de-indentação'.


## Comandos de Repetição

Uma estrutura de repetição, também conhecida como laço (ou loop), é uma ferramenta fundamental em programação que permite a execução de um bloco de código várias vezes, de acordo com uma condição. Em Python, existem duas principais estruturas de repetição: o laço for e o laço while.



### Laço While

O laço while é uma estrutura de repetição baseada em uma condição. Enquanto essa condição for verdadeira, o loop se repetirá.

A sintaxe do laço while é a seguinte:

    while condição:
      executa ação

Esse tipo de laço é útil quando você não sabe de antemão quantas vezes o código deve ser repetido.



````python
contador = 0
while contador < 5:
    print(contador)
    contador += 1
````

Nesse exemplo, o laço while continua executando enquanto contador for menor que 5. A cada iteração, o valor de contador aumenta em 1.

````python
nomes = ["Pedro", "Arthur", "Sofia", "João"]
i = 0

while i < len(nomes):
    print("Olá,", nomes[i])
    i += 1
````

No exemplo acima usamos o laço while com uma lista, percorrendo a lista de nomes usando um índice manual.

Calculando o fatorial de um número usando um laço while:

````python
n = int(input("Digite um número inteiro: "))
f = 1
contador = 1

while contador <= n:
    f *= contador
    contador += 1

print(f"{n}! = {f}")  # essa é uma string formatada. Veremos mais opções de formatação nas próximas aulas.
````

**Atenção!**

O loop while deve ser utilizado com muito cuidado, pois pode resultar em um loop infinito se não houver uma condição de encerramento.



###Controle de laços: break e continue

* **break**: Interrompe o laço antes que ele complete todas as iterações.
* **continue**: Faz com que o laço pule a iteração atual e continue com a próxima.

Podemos parar um loop while com o comando break. Basta utilizar a função dentro do loop. Para melhor utilização, o usuário pode usar uma expressão condicional para determinar onde o loop irá parar.



````python
b = 0
while True:
  b = b + 3
  if b >= 10:
    break
  print(b)

print('Fim')

````

Podemos saltar por valores dentro de um loop while, através do comando continue.


````python
i = 0
while i < 10:
  i += 1   # equivalente a fazer i = i + 1
  if (i+5 < 10):
    continue
  print(i)
````

**Mais exemplos:**

A seguir exemplos de código onde o usuário tem 3 tentativas para acertar uma senha:


````python
senha = "fiscomp1"
tentativas = 0
entrada = ""

while entrada != senha and tentativas < 3:
    entrada = input("Digite a senha: ")
    tentativas += 1

if entrada == senha:
    print("Acesso permitido!")
else:
    print("Acesso negado. Tentativas esgotadas.")
````

````python
senha = "fisicaufrgs"
tentativas = 0
entrada = ""

while tentativas < 3:
    entrada = input("Digite a senha: ")
    if entrada == senha:
        print("Acesso permitido.")
        break
    else:
        print("Senha incorreta.")
        tentativas += 1

if entrada != senha:
    print("Número de tentativas excedido.")
````

Exemplo de código para encontrar o menor número inteiro positivo que seja divisível ao mesmo tempo por 5 e 23

````python
n = 1

while True:     # aqui criamos um laço infinito. Ele vai rodar para sempre, a menos que seja interrompido com break.
    if n % 5 == 0 and n % 23 == 0:
        print(f"O número {n} é divisível por 5 e por 23.")
        break
    n += 1
````

## Laço For

O laço for em Python é geralmente usado para iterar sobre uma sequência (como uma string, lista, tupla, ou range). Ele percorre cada item dessa sequência e executa um bloco de código para cada item.

Esse tipo de estrutura de repetição permite um controle através de uma variável. Essa variável define previamente o número de repetições que o for irá realizar.


Em Python o laço for funciona da seguinte forma:

    for variável in sequência:
      executa ação

onde a sequência pode ser uma string, uma lista ou outro objeto que seja iterável.  

````python
animais = ["gato", "cachorro", "pássaro"]
for bicho in animais:
  print(bicho)
````

Neste exemplo, o laço for percorre a lista animais, imprimindo cada item. A cada 'volta' do loop, a variável bicho recebe um elemento da lista animais.



````python
lista1 = [2.5,7,9,5,12,3]
for j in lista1:
    print(j**2)
````

````python
alunos = ['João', 'Maria', 'José', 'Ana']
for nome in alunos:
  print(nome,len(nome))
````

````python
for letras in 'alfabeto':
    print(letras)
````

Note que a variável não é inicializada antes do loop com um valor e sendo assim, iniciará no primeiro item da sequência e encerrará no último.



Podemos usar os comando break e continue também em loops for.

````python
s = 'flor'
for letra in s:
    if letra == "o":
        break
    print(letra)

print("Fim")
````

````python
for letra in "flor":
    if letra == "o":
        continue
    print(letra)

print("Fim")
````

Podemos usar a cláusula else para loops também!

````python
numeros = [1,2,3,4]
for i in numeros:
    print(i)
else:
    print("isto é escrito já que não há break no for :)")

````

## Função Range()

O loop for também pode ser utilizado para iterar por um intervalo/sequência de números.

Range é uma função que permite criar uma sequência numérica iterável. Range pode ser utilizado de três formas:

- A primeira é através de uma sequência simples, bastando informar qual o valor final da sequência numérica.


    range(fim da sequência numérica)

````python
for j in range(5):
  print(j)
````

No exemplo acima, o início da sequência é o 0 e o fim (5) não é incluído. No entanto, a iteração foi sobre uma sequência de 5 elementos.

- A segunda forma é através da informação do início e do fim da sequência numérica.


    range(início da sequência, fim da sequência)

````python
for k in range(5,11):
  print(k)
````

- A terceira forma é uma variação da segunda, porém, permite definir qual a forma de incremento da lista.


    range(início da sequência, fim da sequência, incremento)

````python
for l in range(3,10,2):
  print(l)
````

````python
for m in range(10, 0, -1):  # podemos contar também de trás para frente usando um incremento negativo
    print(m)
````

Para iterar sobre os índices de uma sequência, podemos combinar as funções range() e len() da seguinte forma:

````python
a = ["gato", "cachorro", "pássaro"]
for i in range(len(a)):
    print(i,a[i])
````

Exemplo: Um possível algoritmo para determinar quais números entre 1 e 10 são pares:


````python
for num in range(1,11): #loop sobre sequência de 1 a 10
    print(num)
    if num % 2 == 0:
        print(num, " é par")
````

Mas veja o que acontece se imprimimos a função range():

````python
print(range(1,6))
````

O que acontece é que o objeto retornado pela função range() se comporta como se fosse uma lista, mas não é uma lista. Na verdade é um objeto iterável, que retorna os itens sucessivos de uma sequência quando você itera sobre a mesma. Se desejarmos transformar em uma lista, devemos usar o list():

````python
list(range(1,6))
````

A função list também converte outros objetos para lista. No exemplo abaixo converte string para lista, onde cada cada caractere da string vira um elemento da lista

## Atividade da última aula: sequência de Fibonacci


Um algoritmo para imprimir os 20 primeiros números da Sequência de Fibonacci (usando loop for)

````python
i = 0     # define os dois primeiros números da sequência
j = 1
print(i)  # imprime os dois primeiros números da sequência i e j
print(j)
contador = 3                # inicializa o contador
while contador <= 20:     # loop para determinar do terceiro ao vigésimo número da sequência
    k = i+j                 # calcula o novo número da sequência k como a soma do último (j) e penúltimo (i) números
    print(k)              # imprime novo número calculado
    i = j                   # atualiza o i
    j = k                   # atualiza o j
    contador += 1           # atualiza o contador
````

Para ajudar na visualização de como o algoritmo funciona

$0 ~1 ~1 ~ 2~ 3 ~5 ~8 ... $

$i ~ ~j ~k$

$~~~ i  ~j ~~k$

$~~~~~ i  ~~j ~~k$
  






Um algoritmo para imprimir os 20 primeiros números da Sequência de Fibonacci (mesmo algoritmo que o acima, só um pouco mais conciso)

````python
i,j = 0,1
print(i)
print(j)
contador = 3
while contador <= 20:
    print(i+j)
    i,j = j,i+j   # atualiza simultaneamente as variáveis i e j
    contador += 1
````

Mesmo algoritmo do exemplo acima, mas agora com um loop for

````python
i,j = 0,1
print(i)
print(j)
for m in range(3,21):
    print(i+j)
    i,j = j,i+j   # atualiza simultaneamente as variáveis i e j
````

Outro algoritmo para imprimir os 20 primeiros números da Sequência de Fibonacci, agora usando listas


````python
l = [0,1]   #adiciona os dois primeiros números da sequência na lista l
contador = 3           # inicializa o contador
while contador <= 20:       # loop para determinar do terceiro ao vigésimo número da sequência
    l.append(l[-1]+l[-2])   # coloca ao final da lista l a soma dos dois últimos elementos dessa lista
   # print(l)
    contador += 1             # atualiza o contador
print(l)
````

Um algoritmo para buscar números primos de 1 até 10 -- para relembrar vários conceitos vistos na última aula


````python
for n in range(1, 11):
    print('n é',n)
    for x in range(2, n):    # loop de 2 a n-1 (não é necessário verificar se n é divisível por 1 ou por ele mesmo)
        print('x é',x)
        if n % x == 0:       # verifica se n é divisível por x
            print(n, 'igual', x, '*', n//x)
            break
    else:
        print(n, 'é um número primo')
````

### Sets

Um set (ou conjunto) é uma coleção não ordenada de valores únicos em Python. Os principais usos incluem testar se um elemento pertence ao conjunto e eliminar entradas duplicadas.

Objetos do tipo set também suportam operações matemáticas de conjuntos, como união, interseção, diferença, etc.

````python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)   # União: {1, 2, 3, 4, 5}
print(a & b)   # Interseção: {3}
print(a - b)   # Diferença: {1, 2}
print(a ^ b)   # Diferença simétrica: {1, 2, 4, 5}
````

## Funções

O Python já possui vários tipos de funções *built-in*, como print(), len(), range(), entre outras. Mas também é possível criar novas funções para executar alguma tarefa. As funções são muito úteis quando precisamos repetir alguma tarefa em diferentes partes do programa. Podemos colocar as instruções para realizar a tarefa em uma função e chamar a função quando for necessário. As funções também ajudam a organizar e modularizar os programas. Em Python as funções são definidas da seguinte forma:

    def nome_da_funcao(argumentos):
      comandos




Em Python, uma função é definida usando a palavra-chave def, seguida pelo nome da função e parênteses. Dentro dos parênteses são definidos os argumentos (parâmetros) que a função recebe como entrada (quando houver). O código que faz parte da função é indentado.



````python
def minha_funcao():
    print('Olá! Esta é a minha função!')
````

No exemplo acima a função minha_funcao não possui nenhum argumento. No corpo da função definimos os comandos que ela irá executar quando for chamada. Para chamar a função, basta fazer

Uma função para escrever sequência de Fibonacci até o número n:

````python
def fib(n):
    """Imprime os números da sequência de Fibonacci ATÉ o número n.

     Isto é uma String de documentação (docstring).
     Algumas ferramentas de documentação utilizam as docstrings para produzir
     automaticamente a documentação online ou permitir que o usuário navegue
     mais facilmente pelo código.

     É uma boa prática de programação incluir sempre docstrings nas suas funções!! ;)

    """
    i, j = 0, 1
    while i < n:
        print(i, end=' ')  #O argumento end pode ser usado para evitar uma nova linha após a saída
        i,j = j,i+j
    return
````

Podemos escrever uma função que devolve uma lista de números da sequência de Fibonacci:

````python
def fib_lista(n):
    """
    Função que retorna uma lista contendo os números da sequência de Fibonacci
    até o número n.
    """
    lista=[]
    i, j = 0, 1
    while i < n:
        lista.append(i)
        i,j = j,i+j

    return lista


f20 = fib_lista(20)
print(f20)
````

## Funções recursivas


A recursividade é um conceito em que uma função chama
a si mesma para resolver um problema.

É necessário ter um caso base para evitar chamadas
infinitas.

````python
def contador_regressivo(n):
  if n == 0:
    print("Fogo!")
  else:
    print(n)
    contador_regressivo(n - 1)


contador_regressivo(3)
````

## Escopo Local e Global

O escopo de uma variável determina onde ela pode ser
acessada dentro do programa.

• Variáveis definidas dentro de uma função têm escopo
local, ou seja, só podem ser acessadas dentro da função.

• Variáveis definidas fora de qualquer função têm escopo
global, podendo ser acessadas em qualquer parte do
programa

````python
a = 5.  # esta é uma variável global
b = 10.    # esta é outra variável global

def func(x):
  #global c
  c = 0.    # esta é uma variável local
  print('Dentro da função:')
  print('a=',a,' b=',b)
  print('c=',c)


print("Antes de iniciar: a=", a,'b=',b)

func(3)   # chamando a função func

print("Após a função:")
print('a=',a,' b=',b)    # a e b ainda existem, pois são globais
print('c=',c)            # Erro: c não existem mais
````

1. Crie uma função em Python para converter coordenadas cartesianas $(𝑥,𝑦)$ em coordenadas polares $(𝑟,\theta)$.

$r = \sqrt{x^2 + y^2}$

$\theta = \rm{arctan} \frac{y}{x}$

Entretanto, quando calculamos a função arco tangente (atan), o resultado está entre $-\pi/2$ e $\pi/2$.

````python
import math
y = -1.
x = -1.
theta = math.atan(y/x)
print(theta*180/math.pi)
````

Então, precisamos ajustar a função para que funcione fora do domínio, cobrindo o intervalo de ângulos completo, ou seja,

\begin{equation}
  \theta =
    \begin{cases}
      \rm{arctan} \frac{y}{x} & \text{se $x > 0$}\\
      \rm{arctan} \frac{y}{x} + \pi & \text{se $x < 0$}\\
    \end{cases}       
\end{equation}

Também precisamos levar em conta os casos em que $x=0$:



*   se $x = 0$ e $y > 0$: $\theta = \pi/2$
*   se $x = 0$ e $y < 0$: $\theta = -\pi/2$




````python
import math

def cartesiana2polar(x,y):
  """Função para converter coordenadas cartesianas (x,y) em coordenadas polares
  (r,theta), onde theta é dado em radianos. """

  r = math.sqrt(x**2+y**2)
  if x > 0:
    theta = math.atan(y/x)
  if x < 0:
    theta = math.atan(y/x) + math.pi
  if x == 0 and y > 0:
    theta = math.pi/2.
  if x == 0 and y < 0:
    theta = -math.pi/2
  if x == 0 and y == 0:
    theta = 0.

  return r,theta
````

````python
print(cartesiana2polar(2,2))
````

Entretanto, o módulo math possui outra função (atan2) que leva em conta os sinais de $x$ e $y$ (ou seja, o quadrante correto) e retorna o ângulo entre  $-\pi$ e $\pi$.

````python
def cartesiana2polar_2(x,y):
  """Função para converter coordenadas cartesianas (x,y) em coordenadas polares
  (r,theta), onde theta é dado em graus. """

  r = math.sqrt(x**2+y**2)
  theta = math.atan2(y,x)

  return r,theta*180./math.pi
````

````python
print(cartesiana2polar_2(-2,-2))
````

## Arrays

Um array é um conjunto ordenado de valores. Mas há diferenças importantes entre listas e arrays:
* O número de elementos de um array é fixo e não é possível adicionar ou remover elementos de um array após criá-lo;
* Os elementos de um array devem ser todos de um mesmo tipo, como float ou inteiros.

As principais vantagens de usar arrays em vez de listas são:
* Arrays podem ter qualquer número de dimensões. Arrays unidimensionais e bidimensionais são, respectivamente, como vetores e matrizes em álgebra linear. Listas são sempre unidimensionais.
* Como se comportam como vetores ou matrizes, podemos fazer qualquer aritmética com eles, ao contrário de listas.
* Arrays funcionam mais rápido que listas em Python.


Listas em Python são mais flexíveis, mas ocupam mais
espaço em memória e têm acesso aos elementos menos
eficiente em comparação com arrays. O acesso aos elementos de uma lista em Python é menos eficiente do que em arrays, principalmente em grandes
quantidades de dados. A diferença de eficiência se deve à forma como as listas e arrays são implementados em nível de memória.




**Quando usar listas em vez de arrays**

Embora os arrays do NumPy sejam mais eficientes para
cálculos numéricos e operações matriciais, há casos em
que as listas podem ser preferíveis.

* Use listas quando precisar de flexibilidade no tamanho
dos dados ou quando os dados não forem
necessariamente numéricos.

* Listas também são úteis quando a manipulação dos
dados requer funcionalidades específicas não disponíveis
em arrays, como adicionar ou remover elementos
dinamicamente.


No entanto, para cálculos numéricos intensivos e
operações matriciais, os arrays do NumPy são altamente
otimizados e oferecem melhor desempenho.

Cada dimensão de um array é chamada axis. O número de axes é chamada de rank do array. A indexação dos arrays também começa em 0, com um número para cada axis do array.

## Exemplo da última aula


Considere um objeto descrevendo um movimento retilínio uniformemente variado (MRUV) ao longo do eixo $x$.
A posição ao longo do eixo $x$ em função do tempo é dada por:

\begin{equation}
x(t) = x_0 + v_0 t + \frac{a t^2}{2}
\end{equation}

e a velocidade em função do tempo é dada por:

$$ v(t) = v_0 + at $$



Criando funções para determinar a posição e velocidade do objeto:


````python
def posicao(t, a, x0=0., v0=0.):
  """Determina a posição de um objeto que descreve MRUV."""
  x = x0 + v0*t + a*t**2/2
  return x
````

````python
def velocidade(t, a, v0=0.):
  """Determina a velocidade de um objeto que descreve MRUV."""
  v = v0 + a*t
  return v
````

Podemos usar as funções para encontrar a posição e velocidade para valores específicos.

Por exemplo, considerando que objeto partiu da origem ($x_0=0$) com velocidade inicial $v_0=0$ e aceleração $a = 5$ m/s$^2$, no tempo $t = 10$ s a sua posição e velocidade serão

````python
t = 10.
a = 5.
print(posicao(t,a))
print(velocidade(t, a))
````

Determinando as posições para $a = 5$ m/s$^2$ e t variando de 0 a 50s em intervalos de 1s usando listas

````python
lista_posicoes = []
for t in range(51):
  lista_posicoes.append(posicao(t,a=5.))

print(lista_posicoes)
````

````python
# podemos converter uma lista para um array
import numpy as np
posicoes_array = np.array(lista_posicoes)
print(posicoes_array)
````

Operações com arrays - **muito mais simples**!!!


````python
import numpy as np
t_a = np.arange(0,51,1)
pos = posicao(t_a, a=5.)
print(t_a)
print(pos)
````

Calculando a posição para tempo de 0 s a 50 s a cada 2.5 s para $a=5$ m/s$^2$, $x_0 = 0$ m $v_0 = 10$ m/s

````python
t = np.arange(0,52.5,2.5)
print(t)
pos = posicao(t, a=5., v0=10.)

print(pos)
````

Determinando a velocidade para tempo de 0 s a 50 s a cada 0.5 s para $a=5$ m/s$^2$, $x_0 = 0$ m $v_0 = 10$ m/s

````python
vel = velocidade(t,a=5., v0=10.)
print(vel)
````

**Operações com arrays**

O poder do Numpy vem da possibilidade de realizar operações com arrays.

Com np.array, as operações são feitas elemento a elemento, sem precisar de loop for:

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)   # Soma: [5 7 9]
print(a - b)   # Subtração: [-3 -3 -3]
print(a * b)   # Multiplicação: [ 4 10 18]
print(b / a)   # Divisão: [4.  2.5 2. ]
````

Operação com escalares:

````python
print(a + 5)    # Soma: [6 7 8]
print(a * 2)    # Multiplicação: [2 4 6]
````

O NumPy fornece funções que atuam sobre arrays inteiros:

````python
print(np.sin(a))
print(np.exp(a))
print(np.sqrt(a))
````

 O NumPy fornece funções que realizam a soma, média e estatísticas dos elementos de um array:

````python
print(a.sum())       # soma dos elementos
print(a.mean())      # média dos elementos
print(a.max())       # maior valor do array
print(a.min())       # menor valor do array
print(a.std())       # desvio padrão dos elementos
````

Funções para realizar operações com matrizes (arrays 2D)


````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[10, 20], [30, 40]])

print(np.dot(A, B)) # Produto matricial
````

Comparações e filtros lógicos

````python
a = np.array([5, 10, 15, 20])

print(a > 10)        # [False False  True  True]
print(a[a > 10])     # filtrando os elementos que são > 10
````

##Gráficos um pouco mais complexos

plt.subplots() é uma função do Matplotlib que retorna uma tupla contendo objetos figura e eixos. Quando usamos fig, ax = plt.subplots() estamos desempacotando esta tupla nas variáveis fig (objeto figura) e ax (objeto eixos).Tendo o objeto figura fig podemos mudar atributos no nível da figura e também salvar a figura como uma imagem depois. Tendo o objeto eixos ax podemos mudar atributos no nível dos eixos.


````python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, f1, '--', color='r', lw=1, ms=4)
ax.text(0.85, 0.7, 'texto aqui', transform=ax.transAxes) ## coordenadas relativas aos Axes da figura
                                                        ## esses vão de (0,0) no canto inferior esquerdo
                                                        ## a (1,1) no canto superior direito
                                                        ## Compare com próximo gráfico
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$\sin(\alpha)$', fontsize=15)     ## Latex para gerar o texto das labels
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, f1, 'o--', color='b', lw=0.6, ms=4)
ax.text(3., 0.75, 'texto aqui', transform=ax.transData)  ## coordenadas relativas aos dados da figura,
                                                        ## controlados por xlim e ylim
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$\sin(\alpha)$', fontsize=15)
plt.show()
````

Múltiplos Axes em uma única figura (vários gráficos em uma mesma figura)

## Histogramas
Histogramas são ferramentas fundamentais de análise em várias áreas do conhecimento. Eles são usados para investigar variáveis ditas aleatórias.
Estas variáveis são associadas a problemas onde a abordagem mais adequada é a estatística. Os histogramas representam as distribuições destas variáveis.



histogram.png

Para criar um histograma devemos determinar o número de ocorrências em cada intervalo (bin). No caso de um histograma de uma variável aleatória real, N(x$_i$) corresponde ao número de ocorrências entre x$_i$ e x$_i$+$\Delta$x, onde $\Delta$x é a largura do bin.
Alternativamente, podemos usar a frequência F(x$_i$) = N(x$_i$)/N$_{total}$.

histogram2.png

Importante notar que N(x$_i$) depende da escolha da largura do intervalo $\Delta$x (ou do número de bins).
histogram3.png

Na Física, os histogramas são amplamente utilizados para analisar a distribuição de diversas grandezas físicas, sendo essenciais para identificar padrões, tendências e anomalias em dados experimentais.

A Mecânica Quântica, que estuda a natureza em nível microscópico, lida com distribuições de variáveis aleatórias em sua formulação de base.

Na Física Clássica, várias problemas são mais facilmente tratados usando-se ferramentas estatísticas.
Por exemplo, para um número suficientemente grande de jogadas de um dado ideal, o número de vezes que se esperam os números 1,2,3,4,5 e 6 deve ser semelhante. O gráfico do número de vezes que cada um desses números ocorreu é um histograma da distribuição de números desse dado.

Qual o resultado esperado? Como fazer esse gráfico?

Iniciamos "jogando" nosso dado.  Para isso podemos usar o módulo random do pacote numpy, que possui uma função chamada randint para gerar um array com números aleatórios

````python
import numpy as np
N = 100000
z = np.random.randint(1,7,N) #a função randint(1,7,N) gera um array com N números aleatórios inteiros no intervalo de interesse.
print(z)  # o array z contém os valores obtidos após N "jogadas"
````

Alternativamente, podemos usar o pacote random que contém uma função também  chamada randint (embora com argumentos diferentes daquela do numpy)


````python
import random as rand
N = 1000
z1 = [rand.randint(1,6) for i in range(N)] #a função randint(1,6) gera os números no intervalo de interesse.
print(z1) # a lista z1 contém os valores obtidos após N "jogadas"
````

Como podemos determinar a qual bin pertence um certo valor do conjunto z?

De forma muito intuitiva, podemos percorrer todos os intervalos e checar se o valor em questão faz parte daquele intervalo.

Ou podemos obter uma maneira mais direta de encontrar qual bin um certo valor pertence.

Para contar a ocorrência de cada número de 1 a 6 da lista z, iniciamos criando um array do numpy com 6 posições correspondendo aos 6 bins (1 para cada valor do dado).

````python
histo = np.zeros(6)
print(histo)
````

ocorrência: 1  2  3  4  5  6

array histo:  0  1  2  3  4  5

 Um algoritmo simples para contar o número de ocorrências de cada valor consiste
 em usar os próprios números da lista z como índices do array histo. A cada ocorrência de um número em z, o valor de histo correspondente ao índice-1 (lembre que os índices começam em zero) é incrementado de 1.


````python
for i in z:
    histo[i-1] = histo[i-1] +1

print(z)
print(histo)
print(np.sum(histo))
````

Fazendo o histograma com barras verticais

````python
import matplotlib.pyplot as plt

x = np.arange(1,7,1) #criando um array com os valores que irão no eixo x no gráfico do histograma

plt.bar(x,histo, color= 'red')
plt.show()
````

````python
plt.hist(z,bins=6,range=(1,7), align='left',color='red', edgecolor= 'black')

plt.show()
````

````python
plt.hist(z, bins=6, edgecolor= 'black')

plt.show()
````

##Uma brevíssima introdução a Classes em Python

Em Python, toda informação que já usamos é representada na forma de um objeto. Por exemplo, o número 1 é um objeto da classe int, 2.75 é um objeto da classe float e assim por diante.

Classes permitem organizar dados e funcionalidades juntos. Criar uma nova classe significa criar um novo “tipo” de objeto, permitindo que novas “instâncias” (exemplos) desse tipo sejam produzidas. Esse mecanismo de classes constitui a chamada **programação orientada a objetos**. Linguagens orientadas a objetos como o Python permitem a definição de novas classes quando se programa.












Classes são abstrações de alguma 'coisa'. Essa 'coisa' possui um estado e um comportamento. Um estado é definido por um conjunto de variáveis, que chamamos de **atributos**. Os estados podem ser modificados por meio de ações sobre o objeto, que definem o seu comportamento. Essas ações correspondem a funções e são chamadas de **métodos**. Então, cada instância de uma classe pode ter atributos anexados a ela ou métodos.

Por exemplo, podemos definir uma classe Veiculo.

Os atributos podem ser:

*   Tipo (avião, carro, barco, ...)
*   Ano
*   Modelo
*   Cor
*   Número de passageiros que comporta
*   velocidade

Os métodos podem ser:

*   acelera(vel): acelera até velocidade vel
*   pare():  faz velocidade = 0




Assim como definições de funções (com comando def), é preciso definir e executar as classes antes para que elas tenham qualquer efeito.

````python
#definição de classe

class Carro:
    pass
````

A classe Carro não possui nenhum atributo nem métodos, mas com ela já podemos criar objetos, ou seja, novas instâncias desta classe.

Para a instanciação de uma classe (“invocar” um objeto classe), usamos a mesma sintaxe de chamar uma função. É como se o objeto classe do exemplo acima fosse uma função sem parâmetros, que devolve uma nova instância da classe.

````python
onix = Carro()
hb20 = Carro()
print(hb20)
````

Além de instanciação, as classes também permitem outro tipo de operação: as referências a atributos. Para criar atributos, basta fazer atribuições usando objeto.atributo e usar os atributos como variáveis.

````python
onix.ano = 2021
onix.cor = 'preto'
onix.modelo = 'Onix'

hb20.ano = 2015
hb20.cor = 'branco'
hb20.modelo = 'HB20'

hb20.ano += 5
print(hb20.ano)
````


A operação de instanciação (“invocar” um objeto classe) acima criou um objeto vazio.  Mas também podemos criar novos objetos com um estado inicial já  pré-determinado. No caso da classe Carro, podemos deixar as propriedades de modelo, ano e cor como atributos da classe e inicializá-las quando um objeto é criado (instanciado). Para isso, usamos na definição da classe um método especial chamado \_\_init\_\_(), conhecido como construtor da classe.





````python
class Carro:

  def __init__(self,m,a,c):
    self.ano = a
    self.cor = c
    self.modelo = m

# Novas instâncias são criadas e nelas os atributos ‘modelo’, ‘ano’ e ‘cor’
# são criados automaticamente com os valores passados como argumentos do construtor

renegade = Carro('Renegade',2021, 'marrom')
t_cross = Carro('T-Cross',2023,'cinza')

````

````python
print(t_cross.modelo)
````

````python
renegade.ano += 2    # podemos acessar e modificar os atributos do objeto renegade
print(renegade.ano)
````

Você deve estar se perguntando o que é esse tal de *self*.

Cada método de uma classe recebe como primeiro argumento uma referência à instância que chama o método. Isso permite que o objeto acesse os seus próprios atributos e métodos. Por convenção, chamamos esse primeiro argumento de self.

Desta forma, ao definirmos qualquer método dentro da classe, um atributo pode ser criado, acessado ou modificado usando self.atributo.



**Métodos**



````python
class Carro:
    def __init__(self, m, a, c, vm=220.):
        self.modelo = m
        self.ano = a
        self.cor = c
        self.vel = 0   # o objeto carro é iniciado com velocidade zero
        self.vmax = vm  # velocidade limite (máxima)


    def acelera(self, v):
        self.vel = v
        if self.vel > self.vmax:
            self.vel = self.vmax
            print( f'Perigo! O {self.modelo} {self.cor} está desgovernado!')
        else:
            print( f'O {self.modelo} {self.cor} está andando a {self.vel:.1f} km/h')


    def para(self):
        self.vel = 0
        print( f'O carro {self.modelo} {self.cor} está parado.')


````

````python
g = Carro('Gol', 1998, 'preto', 180.)   # g é uma instância de Carro
Carro.acelera(g,80.) # chama o método acelera de Carro
````

````python
print(g.vel)
````

O método *acelera* tem dois argumentos: *self* e *v*. Quando o método é executado, *self* é o objeto g e *v* é 80.

````python
Carro.para(g) # chama o método para de Carro
````

````python
print(g.vel)
````

Note que na chamada do método *para* apenas a instância *self* é passada, já que este método só recebe um argumento.

Esse notação deixa explícita a passagem dos objetos como primeiro argumento de cada método, mas ela é redundante, pois todo objeto sabe a que classe ele pertence. (Mas lembre do Zen do Python: 'Explícito é melhor que implícito')

Uma maneira mais enxuta de chamar os métodos é semelhante a dos atributos (usando '.'). Como o primeiro argumento é sempre o próprio objeto, ele pode ser evitado. Por exemplo:

````python
g.acelera(50.)
print(g.vel)
````

Nesse caso em que o objeto é diretamente invocado, o argumento para self é desnecessário e só *v* é passado.


````python
g.para()
print(g.vel)
````

## Caminhante Aleatório
Um modelo físico fundamental na Mecânica Estatística é o do Caminhante Aleatório. Na sua versão mais simples, trata-se de uma partícula movendo-se em um espaço unidimensional que a tempos fixos (por exemplo, t=1,2,3,4,...) com um passo l de tamanho fixo (por exemplo, l=1) com igual probabilidade para um lado ou para o outro nesse espaço unidimensional.

Podemos usar a ideia de classes em Python para explorar o movimento de caminhantes aleatórios.

````python
import random as rand
class RandomWalker:
  x = 0      #todos os caminhantes começam na origem
  def mov(self):
    self.x += 2*rand.randint(0,1)-1
````

````python
#criando uma instância, um elemento, da classe RandomWalker
c1 = RandomWalker()
print(f"Posição inicial de c1, c1.x = {c1.x:2d}")  # verificando a posição

for i in range(10): # movendo c1 10 passos
  c1.mov()
print(f"Posição de c1 após 10 passos c1.x = {c1.x:2d}" )  # verificando a nova posição


#criando o elemento c2
c2 = RandomWalker()
print(f"Posição inicial de c2, c2.x= {c2.x:2d}")  # verificando a posição de c2

for i in range(10): # movendo c2 10 passos
  c2.mov()
print(f"Posição de c2 após 10 passos c2.x = {c2.x:2d}")  # verificando a nova posição de c2
````

````python
#Criando uma lista de M caminhantes
M = 10
c = [RandomWalker() for i in range(M)]

#acessando a posição do quinto elemento da lista
print(f"A posição inicial do elemento 5 da lista é: {c[4].x:2d}")
````

````python
c
````

````python
#Movendo N passos todos os elementos (caminhantes) da lista c
N=200
for i in range(N):
  for j in c:  # loop sobre os caminhantes da lista c
    j.mov()

pos=[c[k].x for k in range(M)]
print(pos)
````

````python
# a mesma coisa de forma mais sintética, usando map e a função lambda
list(map(lambda i:i.mov(), c)) #movendo todos de um passo
pos=list(map(lambda i:i.x,c))  #escrevendo a posição na lista pos
print(pos)
````

````python
#Agora um grande número de caminhantes e muitos passos.
M= 10000  # número de caminhantes
N= 1000 # número de passos

c = [RandomWalker() for i in range(M)]
for i in range(N):
  list(map(lambda i:i.mov(), c))

pos = list(map(lambda i:i.x,c))

#Façamos um histogram da distribuição de posições
import matplotlib.pyplot as plt

plt.hist(pos,bins=30,align='mid')
plt.xlabel('posição do caminhante')
plt.ylabel('N')
plt.show()
````
