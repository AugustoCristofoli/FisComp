# Entradas e saídas

Muitas vezes desejamos ter mais controle sobre a formatação da saída do que simplesmente exibir valores separados por espaço. Existem várias maneiras de formatar a saída. Para mais detalhes e outras opções, consulte https://docs.python.org/pt-br/3/tutorial/inputoutput.html

**Strings formatadas**



Strings literais formatadas (f-strings) permitem incluir o valor de expressões Python dentro de uma string. Para isso é preciso usar f ou F antes da string e escrever expressões na forma {expression}.

```python
cor = 'azul'
print(f'Minha cor favorita é {cor}.')
```

Podemos também escolher o formato que as expressões terão dentro da f-string. Para isso é preciso escrever expressões na forma {expression:formato}.

```python
import math
print(f'O valor de pi é aproximadamente {math.pi:.3f}')  #.3f indica que será um float com 3 casas decimais
```

**Método format()**

O método format atua sobre uma string. A sintaxe é string.format()

```python
print('Minhas cores favoritas são {} e {}.'.format('azul','verde'))
```

```python
# Acessando argumentos pelo nome
print('Coordenadas de Porto Alegre: {longitude}, {latitude}'.format(latitude='-30.0277S', longitude='-51.2287W'))
```

```python
# Acessando argumentos pela posição
s = 'Coordenadas de Porto Alegre: {1}, {0}'
print(s.format('-30.0277S', '-51.2287W'))
```

```python
for x in range(0,20,1):
  print('{0:2d} {1:3d}'.format(x, x**2))
```

```python
print('O valor aproximado de pi é {0:.20f}'.format(math.pi))
```

```python
y = -0.16
print('{0:.1f} {1:.2f} {2:.8f}'.format(y,y**2,y*2))
```

Para mais detalhes e informações: https://docs.python.org/pt-br/3/library/string.html#formatstrings

**Formatação manual**

A função repr() retorna uma string contendo uma representação imprimível de um objeto.
O método string.rjust() justifica uma string à direita, num campo de tamanho definido, acrescentando espaços à esquerda. Similarmente, o método string.ljust() justifica à esquerda e o método string.center() pode ser usado para centralizar.

```python
# mesma tabela de quadrados e cubos anterior, agora formatados manualmente:
for x in range(20):
  print(repr(x).rjust(2), repr(x*x).ljust(3), end=' ')
  print(repr(x**3).center(10))

```

O método string.zfill() preenche uma string numérica com zeros à esquerda. Note que ele lida com sinais positivos e negativos.

```python
s = '-21'
s.zfill(8)  #note que o sinal ocupou uma das posições da string
```

```python
'-21'.zfill(2)
```

```python
'-3.14'.zfill(7) #note que o '.' ocupou uma das posições da string
```

```python
'3.14159'.zfill(2)
```

```python
str(21).ljust(20)
```

A função open() devolve um objeto arquivo (file object) e é frequentemente usada com dois argumentos: open(nome_do_arquivo, modo). O primeiro argumento é uma string contendo o nome do arquivo. * 'r' para apenas ler

* 'a' para abrir para adição

**Numpy.loadtxt()**

Utilizando o método loadtxt do numpy para ler um arquivo de texto com dados. Cada linha do arquivo de texto deve ter a mesma quantidade de valores (tabela).

```python
import numpy as np
col1,col2= np.loadtxt('saida.txt',unpack=True)
print(col1,col2)
```

```python
valores = np.loadtxt('saida.txt',unpack=False)
print(valores)
```

Utilizando o readlines e string.split() para ler o mesmo arquivo:

```python
arquivo = open('saida.txt', 'r')
#criando listas vazias para armazenar os valores das 2 colunas do arquivo
coluna1 = []
coluna2 = []

for line in arquivo:
  coluna1.append(int(line.split(sep=None)[0]))  #separando os elementos da linha
                                                #e colocando o de índice 0
                                                #(transformado para inteiro) na lista coluna1
  coluna2.append(float(line.split(sep=None)[1]))#separando os elementos da linha
                                                #e colocando o de índice 1
                                                #(transformado para float) na lista coluna2

arquivo.close()
print(coluna1,coluna2)
```

Considere um objeto descrevendo um movimento retilínio uniformemente variado (MRUV) ao longo do eixo $x$.

Criando funções para determinar a posição e velocidade do objeto:

```python
def posicao(t, a, x0=0., v0=0.):
  """Determina a posição de um objeto que descreve MRUV."""
  x = x0 + v0*t + a*t**2/2
  return x
```

```python
def velocidade(t, a, v0=0.):
  """Determina a velocidade de um objeto que descreve MRUV."""
  v = v0 + a*t
  return v
```

```python
#Podemos usar as funções para encontrar a posição e velocidade para valores específicos.
#Por exemplo, considerando que objeto partiu da origem (x=0) com velocidade inicial v=0 e
#aceleração a = 5 m/s^2, no tempo t = 10s a sua posição e velocidade serão

t = 10.
a = 5.
print(posicao(t,a))
print(velocidade(t, a))
```

```python
# determinando a posição com arrays
t_a = np.arange(0,21,1)
pos = posicao(t_a, a=5.)

print(pos)
```

```python
# determinando a velocidade com arrays
vel = velocidade(t_a,a=5.)
print(vel)
```
