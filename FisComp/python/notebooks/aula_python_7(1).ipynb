1. Crie uma função em Python para converter coordenadas cartesianas $(𝑥,𝑦)$ em coordenadas polares $(𝑟,\theta)$.

$r = \sqrt{x^2 + y^2}$

$\theta = \rm{arctan} \frac{y}{x}$

Entretanto, quando calculamos a função arco tangente (atan), o resultado está entre $-\pi/2$ e $\pi/2$.

````python
import math
y = -1.
x = -1.
theta = math.atan(y/x)
print(theta*180/math.pi)
````

Então, precisamos ajustar a função para que funcione fora do domínio, cobrindo o intervalo de ângulos completo, ou seja,

\begin{equation}
  \theta =
    \begin{cases}
      \rm{arctan} \frac{y}{x} & \text{se $x > 0$}\\
      \rm{arctan} \frac{y}{x} + \pi & \text{se $x < 0$}\\
    \end{cases}       
\end{equation}

Também precisamos levar em conta os casos em que $x=0$:



*   se $x = 0$ e $y > 0$: $\theta = \pi/2$
*   se $x = 0$ e $y < 0$: $\theta = -\pi/2$




````python
import math

def cartesiana2polar(x,y):
  """Função para converter coordenadas cartesianas (x,y) em coordenadas polares
  (r,theta), onde theta é dado em radianos. """

  r = math.sqrt(x**2+y**2)
  if x > 0:
    theta = math.atan(y/x)
  if x < 0:
    theta = math.atan(y/x) + math.pi
  if x == 0 and y > 0:
    theta = math.pi/2.
  if x == 0 and y < 0:
    theta = -math.pi/2
  if x == 0 and y == 0:
    theta = 0.

  return r,theta
````

````python
print(cartesiana2polar(2,2))
````

Entretanto, o módulo math possui outra função (atan2) que leva em conta os sinais de $x$ e $y$ (ou seja, o quadrante correto) e retorna o ângulo entre  $-\pi$ e $\pi$.

````python
def cartesiana2polar_2(x,y):
  """Função para converter coordenadas cartesianas (x,y) em coordenadas polares
  (r,theta), onde theta é dado em graus. """

  r = math.sqrt(x**2+y**2)
  theta = math.atan2(y,x)

  return r,theta*180./math.pi
````

````python
print(cartesiana2polar_2(-2,-2))
````

# Arrays

Um array é um conjunto ordenado de valores. Mas há diferenças importantes entre listas e arrays:
* O número de elementos de um array é fixo e não é possível adicionar ou remover elementos de um array após criá-lo;
* Os elementos de um array devem ser todos de um mesmo tipo, como float ou inteiros.

As principais vantagens de usar arrays em vez de listas são:
* Arrays podem ter qualquer número de dimensões. Arrays unidimensionais e bidimensionais são, respectivamente, como vetores e matrizes em álgebra linear. Listas são sempre unidimensionais.
* Como se comportam como vetores ou matrizes, podemos fazer qualquer aritmética com eles, ao contrário de listas.
* Arrays funcionam mais rápido que listas em Python.


Listas em Python são mais flexíveis, mas ocupam mais
espaço em memória e têm acesso aos elementos menos
eficiente em comparação com arrays. O acesso aos elementos de uma lista em Python é menos eficiente do que em arrays, principalmente em grandes
quantidades de dados. A diferença de eficiência se deve à forma como as listas e arrays são implementados em nível de memória.




**Quando usar listas em vez de arrays**

Embora os arrays do NumPy sejam mais eficientes para
cálculos numéricos e operações matriciais, há casos em
que as listas podem ser preferíveis.

* Use listas quando precisar de flexibilidade no tamanho
dos dados ou quando os dados não forem
necessariamente numéricos.

* Listas também são úteis quando a manipulação dos
dados requer funcionalidades específicas não disponíveis
em arrays, como adicionar ou remover elementos
dinamicamente.


No entanto, para cálculos numéricos intensivos e
operações matriciais, os arrays do NumPy são altamente
otimizados e oferecem melhor desempenho.

Cada dimensão de um array é chamada axis. O número de axes é chamada de rank do array. A indexação dos arrays também começa em 0, com um número para cada axis do array.


A função **numpy.arange** retorna valores espaçados igualmente em um certo intervalo.
O espaçamento é definido pelo passo (terceiro argumento da função).
```
numpy.arange(início,fim,passo)
```
O valor de início é incluído e o de fim é excluído.

````python
a2 = np.arange(1.,5.5,0.5)
print(a2)
````

A função **numpy.linspace** também retorna n valores espaçados igualmente em um certo intervalo.
```
numpy.linspace(início,fim,n)
```

Nessa função ambos os valores de início e fim são incluídos.

````python
a3 = np.linspace(0,10,5)
print(a3)
````
