## Exemplo da última aula


Considere um objeto descrevendo um movimento retilínio uniformemente variado (MRUV) ao longo do eixo $x$.
A posição ao longo do eixo $x$ em função do tempo é dada por:

\begin{equation}
x(t) = x_0 + v_0 t + \frac{a t^2}{2}
\end{equation}

e a velocidade em função do tempo é dada por:

$$ v(t) = v_0 + at $$



Criando funções para determinar a posição e velocidade do objeto:


```python
def posicao(t, a, x0=0., v0=0.):
  """Determina a posição de um objeto que descreve MRUV."""
  x = x0 + v0*t + a*t**2/2
  return x
```

```python
def velocidade(t, a, v0=0.):
  """Determina a velocidade de um objeto que descreve MRUV."""
  v = v0 + a*t
  return v
```

Podemos usar as funções para encontrar a posição e velocidade para valores específicos.

Por exemplo, considerando que objeto partiu da origem ($x_0=0$) com velocidade inicial $v_0=0$ e aceleração $a = 5$ m/s$^2$, no tempo $t = 10$ s a sua posição e velocidade serão

```python
t = 10.
a = 5.
print(posicao(t,a))
print(velocidade(t, a))
```

Determinando as posições para $a = 5$ m/s$^2$ e t variando de 0 a 50s em intervalos de 1s usando listas

```python
lista_posicoes = []
for t in range(51):
  lista_posicoes.append(posicao(t,a=5.))

print(lista_posicoes)
```

```python
# podemos converter uma lista para um array
import numpy as np
posicoes_array = np.array(lista_posicoes)
print(posicoes_array)
```

Operações com arrays - **muito mais simples**!!!


```python
import numpy as np
t_a = np.arange(0,51,1)
pos = posicao(t_a, a=5.)
print(t_a)
print(pos)
```

Calculando a posição para tempo de 0 s a 50 s a cada 2.5 s para $a=5$ m/s$^2$, $x_0 = 0$ m $v_0 = 10$ m/s

```python
t = np.arange(0,52.5,2.5)
print(t)
pos = posicao(t, a=5., v0=10.)

print(pos)
```

Determinando a velocidade para tempo de 0 s a 50 s a cada 0.5 s para $a=5$ m/s$^2$, $x_0 = 0$ m $v_0 = 10$ m/s

```python
vel = velocidade(t,a=5., v0=10.)
print(vel)
```

# Arrays


Um array é uma estrutura de dados que armazena um conjunto ordenado de elementos, todos do mesmo tipo.

```python
# Criando um array NumPy unidimensional
a = np.array([1, 2, 3, 4])
print(a)

# Operações com arrays
print(a * 2)  # Saída:

# Criando um array multidimensionais (matriz)
b = np.array([[1, 2], [3, 4]])
print(b)
```

**Outras formas de criar arrays**


np.zeros(shape): cria array com todos os elementos iguais a zero, onde shape é o formato do array

```python
a0 = np.zeros(10, dtype = int)  # Array 1D com 10 zeros do tipo inteiro
print(a0)
```

```python
a0_2 = np.zeros((3,2,5)) # Array 3D com tamanho 3x2x5
print(a0_2)
```

np.ones(shape): cria um array com todos os elementos iguais a 1, , onde shape é o formato do array



```python
a1 = np.ones(7) # Array 1D com 7 elementos
print(a1)
```

```python
a1_2 = np.ones((2,3),dtype=int) # Array 2D (matriz) 2x3 com 1s do tipo inteiro
print(a1_2)
```

```python
a1_3 = np.ones((3,2,3,4)) # Array 4D (matriz) 3x2x3x4
print(a1_3)
```

np.full(shape, valor): cria um array preenchido com um valor específico

```python
a3 = np.full((4,2), 3)  # Array 4x2 preenchida com 3
print(a3)
```

np.arange(início, fim, passo): cria um array com valores em uma sequência regular. O valor de início é incluído e o do fim é excluído. É parecida com range() do Python, mas o passo não precisa ser um inteiro.

```python
a4 = np.arange(0, 10, 2.5)
print(a4)
```

np.linspace(início, fim, n): Cria um array com n valores igualmente espaçados entre o início (incluído) e fim (incluído também).

```python
a5 = np.linspace(0, 1, 6)
print(a5)
```

np.eye(n): cria uma matriz identidade de tamanho n x n.

```python
a_i = np.eye(5)

print(a_i)
```

O NumPy também oferece várias funções para criar arrays com números aleatórios, muito úteis em simulações, testes e experimentos numéricos.

Estes métodos estão incluídos no módulo numpy.random.

np.random.rand(d1, d2, ...): cria um array com números aleatórios uniformemente distribuídos no intervalo [0, 1), onde d1, d2, etc são as dimensões


```python
a_r = np.random.rand(2,4,5)  # matriz 2x4x5 com números aleatórios no intervalo [0,1)
print(a_r)
```

np.random.randint(início, fim, size): cria um array com números inteiros aleatórios no intervalo [início, fim) e com tamanho igual a size

```python
a_r2 = np.random.randint(1, 7, size=(2,3))  # matriz 2x3 com inteiros entre 1 e 6
print(a_r2)
```

**Cuidado** ao fazer cópias de arrays

```python
a = np.array([1,1])
print(a)
b = a
print(b == a)
a[0] = 2
print(a)
print(b)
```

```python
a is b
```

```python
a = np.array([1,1])
b = np.copy(a)  # np.copy cria uma cópia do array
a[0] = 2
print(a)
print(b)
```

```python
a is b
```

**Operações com arrays**

O poder do Numpy vem da possibilidade de realizar operações com arrays.

Com np.array, as operações são feitas elemento a elemento, sem precisar de loop for:

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)   # Soma: [5 7 9]
print(a - b)   # Subtração: [-3 -3 -3]
print(a * b)   # Multiplicação: [ 4 10 18]
print(b / a)   # Divisão: [4.  2.5 2. ]
```

Operação com escalares:

```python
print(a + 5)    # Soma: [6 7 8]
print(a * 2)    # Multiplicação: [2 4 6]
```

O NumPy fornece funções que atuam sobre arrays inteiros:

```python
print(np.sin(a))
print(np.exp(a))
print(np.sqrt(a))
```

 O NumPy fornece funções que realizam a soma, média e estatísticas dos elementos de um array:

```python
print(a.sum())       # soma dos elementos
print(a.mean())      # média dos elementos
print(a.max())       # maior valor do array
print(a.min())       # menor valor do array
print(a.std())       # desvio padrão dos elementos
```

Funções para realizar operações com matrizes (arrays 2D)


```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[10, 20], [30, 40]])

print(np.dot(A, B)) # Produto matricial
```

Comparações e filtros lógicos

```python
a = np.array([5, 10, 15, 20])

print(a > 10)        # [False False  True  True]
print(a[a > 10])     # filtrando os elementos que são > 10
```
